```python
#import packages
import pandas as pd
import numpy as np
```

```python
#importing data
orders = pd.read_csv('fact_order_lines.csv')
order_agg = pd.read_csv('fact_orders_aggregate.csv')
order_target = pd.read_csv('dim_targets_orders.csv')
```

```python
#Finding how many Orders we got.

no_of_orders = orders['order_id'].nunique()
no_of_orders
```

```python
#Finding Total Order Lines.
total_order_lines = orders['order_id'].count()
total_order_lines
```

```python
#Finding Line Fill Rate
LIFR = sum(orders['In Full']==1) / total_order_lines
LIFR
```

```python
# Finding Volume Fill Rate
VOFR = sum(orders['delivery_qty']) / sum(orders['order_qty'])
VOFR
```

```python
#Finding On Time Delivery %
OT = sum(order_agg['on_time'] ==1)/ no_of_orders
OT
```

```python
#Finding In Full Delivery %
IF = sum(order_agg['in_full'] ==1)/ no_of_orders
IF
```

```python
#Finding On time & In Full Delivery %
OTIF = sum(order_agg['otif'] ==1)/ no_of_orders
OTIF
```

```python
from numpy.lib.function_base import average
# Finding Average targets from target_oders

# Average On Time Target
avg_OT_tar = average(order_target['ontime_target%'])
print(avg_OT_tar)
# Average In Full Target
avg_IF_tar = average(order_target['infull_target%'])
print(avg_IF_tar)
# Average On Time In Full Target
avg_OTIF_tar = average(order_target['otif_target%'])
print(avg_OTIF_tar)
```

```python
# Required Metrics

print('1. Total order lines =', total_order_lines)
print('2. Line Fill Rate =', round(LIFR,2))
print('3. Volume fill Rate =', round(VOFR,2))
print('4. No of Orders =', no_of_orders)
print('5. On Time delivery % =', round(OT,2))
print('6. In Full Delivery % =', round(IF,2))
print('7. On Time & In Full Delivery =', round(OTIF,2))
print('8. Average On Time delivery % target = ',round(avg_OT_tar,2))
print('9. Average In Full delivery % target = ',round(avg_IF_tar,2))
print('10. Average On Time & Full Time delivery % target = ',round(avg_OTIF_tar,2))
```
